```python
merged_df.head()
```

```python
merged_df_cleaned = merged_df[['age_group','country','sharePopinAgeGroup','shareMEPSinAgeGroup']]
merged_df_cleaned_melted = pd.melt(merged_df_cleaned,id_vars=['age_group','country'])
```

```python
merged_df_cleaned_melted.head()
```

```python
merged_df_cleaned_melted.columns=(['age_group','country','group','share'])
```

```python
countries = set(merged_df.country.to_list())

for country in countries:
    tempdf = merged_df_cleaned_melted[merged_df_cleaned_melted['country']==country]
    
    tempdf_all = tempdf[tempdf['group']=='sharePopinAgeGroup']
    tempdf_mep = tempdf[tempdf['group']=='shareMEPSinAgeGroup']
    
    chart_all = alt.Chart(tempdf_all).mark_bar(color='red', opacity=0.5).encode(
        alt.Y('share:Q',scale = alt.Scale(domain=[0,50]), title=''),
        alt.X('age_group:N', title=''),
    ).properties(width=100,height=100)
    
    
    chart_mep = alt.Chart(tempdf_mep).mark_bar(color='blue', opacity=0.5).encode(
        alt.Y('share:Q',scale = alt.Scale(domain=[0,50]), title=''),
        alt.X('age_group:N',title='')
    ).properties(width=100,height=100, title=country)
    
    both = chart_all+chart_mep
    
    both.display()
```
